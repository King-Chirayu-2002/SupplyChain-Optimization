```python
# Summing up various cost-related fields to get a total cost
df_cleaned['Total Cost'] = (df_cleaned['Order Item Product Price'] * df_cleaned['Order Item Quantity']) - df_cleaned['Order Item Discount'] + df_cleaned['Order Profit Per Order']
```

```python
# Function to calculate distance using latitude and longitude (Haversine formula)
def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # Earth radius in kilometers
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat / 2.0)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2.0)**2
    c = 2 * np.arcsin(np.sqrt(a))
    return R * c
```

```python
import numpy as np
```

```python
# Assuming the shipment starts from a fixed point (e.g., warehouse) for simplicity
warehouse_lat = df_cleaned['Latitude'].mean()
warehouse_lon = df_cleaned['Longitude'].mean()

# Calculate the distance from the warehouse to the delivery location
df_cleaned['Distance'] = haversine(warehouse_lat, warehouse_lon, df_cleaned['Latitude'], df_cleaned['Longitude'])

# Estimating CO2 emissions based on Shipping Mode and Distance
emission_factors = {
    'Road/Water': 15,  # g CO2 per km per tonne for water
    'Air/Road': 500    # g CO2 per km per tonne for air
}

df_cleaned['CO2 Emissions'] = df_cleaned.apply(lambda row: emission_factors[row['Transport Method']] * row['Distance'], axis=1)
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
X = df_cleaned[['Total Cost', 'Shipping Time', 'CO2 Emissions']].values

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Define number of clusters
n_clusters = 3

# Fit K-Means
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
df_cleaned['Cluster'] = kmeans.fit_predict(X_scaled)

# Calculate the average objective function per cluster
df_cleaned['Objective Function'] = (df_cleaned['Total Cost'] + df_cleaned['Shipping Time'] + df_cleaned['CO2 Emissions']) / 3
cluster_means = df_cleaned.groupby('Cluster')['Objective Function'].mean()

# Find the cluster with the minimum average objective function
min_cluster = cluster_means.idxmin()
print(f"Cluster with minimum average objective function: {min_cluster}")
```

```python
# Extract the data points belonging to cluster 0
cluster_0_data = df_cleaned[df_cleaned['Cluster'] == 0]

# Analyze the average values of the features in cluster 0
cluster_0_mean = cluster_0_data[['Total Cost', 'Shipping Time', 'CO2 Emissions']].mean()
print(f"Average features in cluster 0:\n{cluster_0_mean}")
```

```python
import matplotlib.pyplot as plt

# Plot the features of cluster 0
plt.figure(figsize=(10, 6))
plt.scatter(cluster_0_data['Total Cost'], cluster_0_data['Shipping Time'], c='blue', label='Cluster 0')
plt.xlabel('Total Cost')
plt.ylabel('Shipping Time')
plt.title('Cluster 0: Total Cost vs. Shipping Time')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import silhouette_score

silhouette_avg = silhouette_score(X_scaled, df_cleaned['Cluster'])
print(f"Silhouette Score: {silhouette_avg}")
```
